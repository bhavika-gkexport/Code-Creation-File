```python
df = pd.read_excel("Merge_T (1).xlsx")

# Concatenate all sheets into one DataFrame
# combined_df = pd.concat(df.values(), ignore_index=True)
combined_df = df

gemstonetype = []
stoneshape = []
gemstonequality = []
gemstonegrade = []
gemstonesize = []
facetedcabochon = []
gemstonepr = []
perpcorpercarat = []
box_no = []
item_code = []
for i in combined_df.values.tolist():
    if i[3] in ['Gemstone Type','Gemstone Type1']:
        gemstonetype.append(i[4])
        box_no.append(i[1])
        item_code.append(i[0])
    if i[3] == 'Stone Shape':
        stoneshape.append(i[4])
    if i[3] == 'Gemstone Quality':
        gemstonequality.append(i[4])
    if i[3] == 'Gemstone Grade':
        gemstonegrade.append(i[4])
    if i[3] == 'Gemstone Size':
        gemstonesize.append(i[4])
    if i[3] == 'Faceted/Cabochon':
        facetedcabochon.append(i[4])
    if i[3] == 'Gemstone PR':
        gemstonepr.append(i[4])
    if i[3] == 'Per Pc or Per Carat':
        perpcorpercarat.append(i[4])

# Creating a new DataFrame
data = {
    'Item Code':item_code,
    'Gemstone Type': gemstonetype,
    'Stone Shape': stoneshape,
    'Gemstone Quality': gemstonequality,
    'Gemstone Grade': gemstonegrade,
    'Gemstone Size':gemstonesize,
    'Faceted/Cabochon':facetedcabochon,
    'Gemstone PR':gemstonepr,
    'Per Pc or Per Carat':perpcorpercarat,
    'Box No':box_no
}
new_df = pd.DataFrame(data)
# new_df.to_excel("Final Open Diamond With Box before 19R.xlsx",index=False)
```
